```python
import os
from pathlib import Path

import polars as pl

data_dir = Path("../data")
data_dir = data_dir / "AAAI2023/all"
data_dir.mkdir(exist_ok=True, parents=True)

db_url = os.getenv("DATABASE_URL", "postgresql://postgres:postgres@scraper-db:5432/dev")
```

Read from PostgreSQL database

```python
item_query = "select * from conferenceitem where conference like 'AAAI'"
author_query = "select * from author"
keyword_query = "select * from keyword"
author_item_rel_query = "select * from authorpublink"
kw_item_rel_query = "select * from keywordpublink"

items = pl.read_sql(item_query, db_url)
authors = pl.read_sql(author_query, db_url)
keywords = pl.read_sql(keyword_query, db_url)
author_item_relations = pl.read_sql(author_item_rel_query, db_url)
keyword_item_relations  = pl.read_sql(kw_item_rel_query, db_url)
```

Write to parquet

```python
items.write_parquet(data_dir / "items.parquet")
authors.write_parquet(data_dir / "authors.parquet")
keywords.write_parquet(data_dir / "keywords.parquet")
author_item_relations.write_parquet(data_dir / "author_item_relations.parquet")
keyword_item_relations.write_parquet(data_dir / "keyword_item_relations.parquet")
```
